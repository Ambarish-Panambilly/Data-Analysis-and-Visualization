## Titanic passenger Data Visualization in Seaborn

In this notebook, I will try to visualize the famous Titanic passenger dataset using the popular Seaborn visualization library. 

### Basic Imports 

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

### The Data 

```python
titanic_data = sns.load_dataset('titanic')
```

```python
titanic_data.head()
```

### Joint plot

```python
sns.jointplot(x='fare',y='age',data=titanic_data)
```

### Distribution plot 

```python
sns.distplot(titanic_data['fare'],bins=40,kde=False,color='red')
```

### Box plot

```python
sns.boxplot(x='class',y='age',data=titanic_data,palette='cubehelix')
```

### Swarm plot 

```python
sns.swarmplot(x='class',y='age',data=titanic_data,palette='viridis')
```

### Count plot 

```python
sns.countplot(x='sex',data=titanic_data)
```

### Heatmap

```python
sns.heatmap(titanic_data.corr(),cmap='coolwarm')
plt.title('titanic_data.correlation')
```

### Facetgrid

```python
g = sns.FacetGrid(data=titanic_data,col='sex')
g.map(plt.hist,'age')
```
